```python
my_title = 'TITLE'
my_xlabel = 'X axis_LABEL'
my_ylabel = 'Y axis_LABEL'

ax = df['Close'].plot(figsize=(12,6), title=my_title)
ax.set(xlabel=my_xlabel, ylabel=my_ylabel)
```

```python
ax = df['Close'].plot(figsize=(12,6), title=my_title)
ax.autoscale(tight=True, axis='both')
ax.set(xlabel=my_xlabel, ylabel=my_ylabel)
```

```python
ax = df['Close']['2017-01-01':'2017-12-31'].plot(figsize=(12,6),title=my_title)
ax.set(xlabel=my_xlabel, ylabel=my_ylabel)
```

```python
ax = df['Close'].plot(figsize=(12,6), xlim=['2017-01-01','2017-12-31'], title=my_title)
ax.set(xlabel=my_xlabel, ylabel=my_ylabel)
```

```python
ax = df['Close'].plot(figsize=(12,6), xlim=['2017-01-01','2017-12-31'], ylim=[0,70], title=my_title)
ax.set(xlabel=my_xlabel, ylabel=my_ylabel)
```

```python
ax = df['Close']['2017-01-01':'2017-12-31'].plot(figsize=(12,6),title=my_title,ls='--',color='green')
ax.set(xlabel=my_xlabel, ylabel=my_ylabel)
```

### Spacing for x-ticks

```python
from matplotlib import dates
```

```python
df['Close'].plot(figsize=(12,6), xlim=['2017-01-01','2017-03-01'], ylim=[50,60])
```

```python
ax = df['Close'].plot(figsize=(12,6), xlim=['2017-01-01','2017-03-01'], ylim=[50,60])
ax.set(xlabel='')
ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))
```

```python
ax = df['Close'].plot(figsize=(12,6), xlim=['2017-01-01','2017-03-01'], ylim=[50,60])
ax.set(xlabel='')
ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_major_formatter(dates.DateFormatter("%a-%B-%d"))
```

```python
ax = df['Close'].plot(figsize=(12,6), xlim=['2017-01-01','2017-03-01'], ylim=[50,60])
ax.set(xlabel='')
ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_major_formatter(dates.DateFormatter("%d"))

ax.xaxis.set_minor_locator(dates.MonthLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter("%b"))
```

```python
ax = df['Close'].plot(figsize=(12,6), xlim=['2017-01-01','2017-03-01'], ylim=[50,60])
ax.set(xlabel='')
ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_major_formatter(dates.DateFormatter("%d"))

ax.xaxis.set_minor_locator(dates.MonthLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter("\n\n%b"))
```

```python
ax = df['Close'].plot(figsize=(12,6), xlim=['2017-01-01','2017-03-01'], ylim=[50,60])
ax.set(xlabel='')
ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_major_formatter(dates.DateFormatter("%d"))

ax.xaxis.set_minor_locator(dates.MonthLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter("\n\n%b"))

ax.yaxis.grid(True)
ax.xaxis.grid(True)
```
